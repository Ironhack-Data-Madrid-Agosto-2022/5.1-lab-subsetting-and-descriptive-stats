#### Find the minimum, mean and the maximum of all numeric columns for each Department.

##### Hint: Use functions from Data Manipulation lesson

```python
employees = employee.groupby('Department', as_index = False).agg({'Salary': [np.min, np.mean, np.max], 'Years': [np.min, np.mean, np.max]})
employees.columns = ['Department', 'Salary_min','Salary_mean','Salary_max','Years_min', 'Years_mean', 'Years_max']
employees
```

### Bonus Question

#### For each department, compute the difference between the maximal salary and the minimal salary.

##### * Hint: try using `agg` or `apply` and `lambda`*

```python
employees = employee.groupby('Department', as_index = False).agg({'Salary': lambda x: max(x)-min(x)})
employees.columns = ['Department', 'Diferent']
employees
```
